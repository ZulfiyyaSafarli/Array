```python
#https://leetcode.com/problems/two-sum/
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        dictionary = {}
        for i in range(n):
            if(target - nums[i]) in dictionary:
                return [dictionary[target - nums[i]], i]
            dictionary[nums[i]] = i
```

```python
#https://leetcode.com/problems/how-many-numbers-are-smaller-than-the-current-number/
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        empty_dict = {}
        empty_list = []
        n = len(nums)

        # count frequency of each number
        for num in nums:
            if num in empty_dict:
                empty_dict[num] += 1
            else:
                empty_dict[num] = 1

        # sort the keys
        sorted_keys = sorted(empty_dict.keys())

        # map each number to how many are smaller than it
        smaller_count = {}
        count = 0
        for key in sorted_keys:
            smaller_count[key] = count
            count += empty_dict[key]

        # build the answer list
        for num in nums:
            empty_list.append(smaller_count[num])

        return empty_list
```

```python
#https://leetcode.com/problems/length-of-last-word/
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        end = len(s) - 1
        while s[end] == " ":
            end -= 1

        start = end
        while start >= 0 and s[start] != " ":
            start -= 1

        return end - start
```

```python
# https://leetcode.com/problems/move-zeroes/
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        n = len(nums)
        left = 0
        for right in range(n):
            if nums[right] != 0:
                nums[right], nums[left] = nums[left], nums[right]
                left += 1

        return nums
```

```python
# https://leetcode.com/problems/remove-duplicates-from-sorted-array/
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if not nums:
            return 0

        left = 1
        for right in range(1, len(nums)):
            if nums[right] != nums[right - 1]:
                nums[left] = nums[right]
                left += 1

        return left
```

```python
#https://leetcode.com/problems/sorting-the-sentence/
class Solution:
    def sortSentence(self, s: str) -> str:
        words = s.split()
        # create empty list of correct size
        result = [None] * len(words)
        for word in words:
            # Convert the last character (digit) to a 0-based index
            # ord('1') = 49 → 49 - 49 = 0
            # ord('2') = 50 → 50 - 49 = 1
            index = ord(word[-1]) - 49
            clean_word = word[:-1]    # remove last char
            result[index] = clean_word

        return " ".join(result)
```

```python
#https://leetcode.com/problems/find-words-that-can-be-formed-by-characters/
class Solution:
    def countCharacters(self, words: List[str], chars: str) -> int:
        # Count available letters
        letters = {}
        for c in chars:
            if c in letters:
                letters[c] += 1
            else:
                letters[c] = 1

        res = 0

        for word in words:
            # Reset dictionary for each word
            word_letters = {}

            # Count letters in the current word
            for c in word:
                if c in word_letters:
                    word_letters[c] += 1
                else:
                    word_letters[c] = 1

            # Check if the word can be formed
            valid = True
            for c in word_letters:
                if c not in letters or word_letters[c] > letters[c]:
                    valid = False
                    break

            # Add length if valid
            if valid:
                res += len(word)

        return res
```

```python
##https://leetcode.com/problems/find-words-that-can-be-formed-by-characters/
class Solution(object):
    def countCharacters(self, words, chars):
        # Create a list of size 26 (for each letter 'a' to 'z') initialized to 0
        # This will store the frequency of each character in `chars`
        counts = [0] * 26

        # Step 1: Count how many times each character appears in `chars`
        for ch in chars:
            # ord(ch) gives the ASCII code of the character
            # ord('a') is subtracted to map 'a'->0, 'b'->1, ..., 'z'->25
            counts[ord(ch) - ord('a')] += 1

        # Initialize the result to store the total length of words that can be formed
        result = 0

        # Step 3: Iterate over each word in the list of words
        for word in words:
            # Check if the current word can be formed using `chars`
            if self.canForm(word, counts):
                # Step 4: If it can be formed, add its length to the result
                result += len(word)

        # Return the total sum of lengths of all words that can be formed
        return result

    def canForm(self, word, counts):
        # Create a list of size 26 to count frequency of characters in `word`
        c = [0] * 26

        # Step 2: Count characters in the current word
        for ch in word:
            # Map character to index 0-25
            x = ord(ch) - ord('a')
            # Increment the count for this character
            c[x] += 1
            # If the word uses more of a character than available in `chars`, return False
            if c[x] > counts[x]:
                return False

        # If all character counts are within limits, return True
        return True
```

```python
#https://leetcode.com/problems/last-day-where-you-can-still-cross/
from typing import List

class Solution:
    def frequencySort(self, nums: List[int]) -> List[int]:
        # Step 1: Count frequencies
        freq = {}
        for number in nums:
            if number in freq:
                freq[number] += 1
            else:
                freq[number] = 1

        # Step 2: Sort keys by frequency ascending, but by number descending if equal frequency
        sorted_keys = sorted(freq, key=lambda x: (freq[x], -x))
        # sorted(freq, ...) is essentially sorting the keys of freq.
        # key tells Python what value to use when comparing elements during sorting.
        # x is each key in the dictionary.
        # lambda x: (freq[x], -x) returns a tuple for each key: (frequency, negative_number).
        # freq[x] → the frequency of the key x: Sorting is done first by frequency in ascending order.
        # -x → negative of the key: If two keys have the same frequency, the one with the larger number comes first.

        # Step 3: Build the result list
        result = []
        for key in sorted_keys:
            result.extend([key] * freq[key])

        return result
```

```python
# https://leetcode.com/problems/sort-array-by-increasing-frequency/
class Solution:
    def frequencySort(self, nums: List[int]) -> List[int]:
        # Create a list of size 201 to count frequencies for numbers from -100 to 100
        # Index 0 corresponds to -100, index 200 corresponds to 100
        freq = [0] * 201

        # Step 1: Fill frequency array
        for x in nums:
            # Increment frequency for number x
            # We do x+100 to shift negative numbers to positive indices
            # Subtracting 1 to make smaller frequencies "more negative" for sorting
            freq[x + 100] -= 1

        # Step 2: Sort nums using a custom key
        # key=lambda x: (freq[x+100], x)
        # - freq[x+100] → primary key: smaller frequency comes first
        # - x → secondary key: original number
        # reverse=True → because freq values are negative, we reverse to get correct order
        return sorted(nums, key=lambda x: (freq[x + 100], x), reverse=True)
```

```python
# https://leetcode.com/problems/sum-of-unique-elements/
class Solution:
    def sumOfUnique(self, nums: List[int]) -> int:
        # Step 1: Count frequency of each number
        unique = {}
        for n in nums:
            if n in unique:
                unique[n] += 1  # increment count if number already seen
            else:
                unique[n] = 1   # initialize count to 1 if first time seen

        # Step 2: Initialize sum of unique numbers
        summa = 0

        # Step 3: Iterate over each key-value pair in the dictionary
        for key, value in unique.items():
            # Only add the number to sum if it occurs exactly once
            if value == 1:
                summa += key

        # Step 4: Return the total sum of unique numbers
        return summa
```

*   Last position (i = len(flowerbed)-1): Can plant if the previous spot is empty (flowerbed[i-1] != 1)

*   Middle positions: Can plant if both neighbors are empty (flowerbed[i-1] != 1 and flowerbed[i+1] != 1)

```python
#https://leetcode.com/problems/can-place-flowers/
class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        count = 0
        for i in range(len(flowerbed)):
            if flowerbed[i] == 0:
                if (i == 0 and (len(flowerbed) == 1 or flowerbed[i + 1] != 1)):
                    flowerbed[i] = 1
                    count += 1

                elif i == len(flowerbed)-1 and flowerbed[i-1] != 1:
                    flowerbed[i] = 1
                    count += 1

                elif flowerbed[i-1] != 1 and  flowerbed[i+1] != 1:
                    flowerbed[i] = 1
                    count += 1

            if count == n:
                return True

        return count >= n
```

*   Loop through the array from start to end, break if a pair violates increasing order; if you reach the end, the array is monotone increasing.
*   Loop again, break if a pair violates decreasing order; if you reach the end, the array is monotone decreasing.
*   If neither loop reaches the end, the array is not monotone.

```python
#https://leetcode.com/problems/monotonic-array/
class Solution:
    def isMonotonic(self, nums: List[int]) -> bool:
        n = len(nums)
        if n == 1: return True

        # check if this is increasing
        for i in range(1, n):
            if nums[i] < nums[i-1]:
                break

            if i == n - 1:
                return True

        # check if this is decreasing
        for i in range(1, n):
            if nums[i] > nums[i-1]:
                break

            if i == n - 1:
                return True

        return False
```

```python
#https://leetcode.com/problems/baseball-game/
class Solution:
    def calPoints(self, operations: List[str]) -> int:
        stack = []

        for op in operations:
            if op == "C":
                stack.pop()
            elif op == "D":
                stack.append(stack[-1] * 2)
            elif op == "+":
                stack.append(stack[-1] + stack[-2])
            else:
                stack.append(int(op))

        return sum(stack)
```

```python
#https://leetcode.com/problems/find-the-winner-of-the-circular-game/
class Solution:
    def findTheWinner(self, n: int, k: int) -> int:
        # Initialize the winner index
        # We use 0-based indexing here, so initially the first person (index 0) is "winner"
        winner = 0

        # Loop through players from 2 to n
        # We calculate the position of the winner step by step
        for current_number_of_players in range(2, n + 1):
            # The new winner is (previous winner + k) % number of players
            winner = (winner + k) % current_number_of_players

        # Convert from 0-based index to 1-based index
        return winner + 1
```

```python
from collections import deque
from typing import List

class Solution:
    def latestDayToCross(self, row: int, col: int, cells: List[List[int]]) -> int:

        # Helper: convert 1-based to 0-based
        def convert_to_grid(r, c):
            return r - 1, c - 1

        # Helper: check if a path exists using BFS
        def is_path_exists(grid):
            directions = [(-1,0), (1,0), (0,-1), (0,1)]
            queue = deque()
            visited = [[False]*col for _ in range(row)]

            # Start BFS from all land cells in top row
            for c_idx in range(col):
                if grid[0][c_idx] == 0:
                    queue.append((0, c_idx))
                    visited[0][c_idx] = True

            while queue:
                r_idx, c_idx = queue.popleft()
                if r_idx == row - 1:  # reached bottom
                    return True
                for dr, dc in directions:
                    nr, nc = r_idx + dr, c_idx + dc
                    if 0 <= nr < row and 0 <= nc < col and grid[nr][nc] == 0 and not visited[nr][nc]:
                        queue.append((nr, nc))
                        visited[nr][nc] = True
            return False

        # Helper: flood the grid up to 'day' days
        def flood_grid(day):
            grid = [[0]*col for _ in range(row)]
            for i in range(day):
                r, c = convert_to_grid(cells[i][0], cells[i][1])
                grid[r][c] = 1
            return grid

        # Binary search for the last day
        left, right = 0, len(cells)
        last_day = 0

        while left <= right:
            mid = (left + right) // 2
            grid = flood_grid(mid)

            if is_path_exists(grid):
                last_day = mid
                left = mid + 1
            else:
                right = mid - 1

        return last_day
```

```python
class Solution:
    def minimumDeleteSum(self, s1: str, s2: str) -> int:
        # Step 1: Get lengths of the strings
        n = len(s1)
        m = len(s2)

        # Step 2: Create DP table
        # dp[i][j] = maximum ASCII sum of common subsequence
        # using first i characters of s1 and first j characters of s2
        dp = []
        for i in range(n + 1):
            row = []
            for j in range(m + 1):
                row.append(0)
            dp.append(row)

        # Step 3: Fill the DP table
        for i in range(n):
            for j in range(m):
                # If current characters match, keep them
                if s1[i] == s2[j]:
                    previous_value = dp[i][j]
                    ascii_value = ord(s1[i])
                    dp[i + 1][j + 1] = previous_value + ascii_value
                else:
                    # If characters don't match, take the maximum of:
                    # 1. Skipping s1[i]
                    # 2. Skipping s2[j]
                    value_skip_s1 = dp[i][j + 1]
                    value_skip_s2 = dp[i + 1][j]
                    if value_skip_s1 > value_skip_s2:
                        dp[i + 1][j + 1] = value_skip_s1
                    else:
                        dp[i + 1][j + 1] = value_skip_s2

        # Step 4: Calculate total ASCII sum of all characters in s1
        total_ascii_s1 = 0
        for ch in s1:
            total_ascii_s1 += ord(ch)

        # Step 5: Calculate total ASCII sum of all characters in s2
        total_ascii_s2 = 0
        for ch in s2:
            total_ascii_s2 += ord(ch)

        total_ascii = total_ascii_s1 + total_ascii_s2

        # Step 6: Calculate minimum delete sum
        max_common_ascii = dp[n][m]
        min_delete_sum = total_ascii - 2 * max_common_ascii

        # Step 7: Return the answer
        return min_delete_sum
```

```python
from typing import List
from collections import defaultdict
from itertools import pairwise

class Solution:
    def separateSquares(self, squares: List[List[int]]) -> float:
        total_area = 0  # Initialize total area of all squares
        diff = defaultdict(int)  # Dictionary to store sweep line differences

        # Loop through all squares
        for _, y, l in squares:
            total_area += l * l  # Add area of current square to total
            diff[y] += l         # Start of the square: add its width to diff
            diff[y + l] -= l     # End of the square: subtract its width from diff

        area = 0  # Accumulated area while sweeping
        s = 0     # Active width at current y level

        # Loop through consecutive y-values in sorted order
        for y, y2 in pairwise(sorted(diff)):
            s += diff[y]  # Update active width by adding difference at current y
            area += s * (y2 - y)  # Add area of horizontal strip (width * height)

            # Check if accumulated area has reached half of total area
            if area * 2 >= total_area:
                # Interpolate within current strip to find exact y where half-area is reached
                return y2 - (area * 2 - total_area) / (s * 2)
```

```python
#https://leetcode.com/problems/two-sum/
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        dictionary = {}
        for i in range(n):
            if(target - nums[i]) in dictionary:
                return [dictionary[target - nums[i]], i]
            dictionary[nums[i]] = i
```

```python
#https://leetcode.com/problems/concatenation-of-array/
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        n = len(nums)
        empty_list = []
        for i in range(n):
            empty_list.append(nums[i])
        for i in range(n):
            empty_list.append(nums[i])

        return empty_list
```

```python
#https://leetcode.com/problems/running-sum-of-1d-array/
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        empty_list = []
        n = len(nums)
        summa = 0
        for i in range(n):
            summa = summa + nums[i]
            empty_list.append(summa)

        return empty_list
```

```python
#https://leetcode.com/problems/richest-customer-wealth/
class Solution:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        empty_list = []
        rows = len(accounts)
        cols = len(accounts[0])
        for i in range(rows):
            summa = 0
            for j in range(cols):
                summa = summa + accounts[i][j]
            empty_list.append(summa)
        maxi = 0
        for i in range(len(empty_list)):
            if empty_list[i] > maxi:
                maxi = empty_list[i]
        return maxi
```

```python
#https://leetcode.com/problems/kids-with-the-greatest-number-of-candies/
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        n = len(candies)
        maxi = 0
        empty_list = []
        for i in range(n):
            if candies[i] > maxi:
                maxi = candies[i]

        for i in range(n):
            if candies[i] + extraCandies >= maxi:
                empty_list.append(True)
            else:
                empty_list.append(False)

        return empty_list
```

```python
#https://leetcode.com/problems/shuffle-the-array/
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        empty_list = []
        n = len(nums) // 2
        for i in range(n):
            empty_list.append(nums[i])
            empty_list.append(nums[i+n])

        return empty_list
```

```python
#https://leetcode.com/problems/how-many-numbers-are-smaller-than-the-current-number/
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        empty_dict = {}
        empty_list = []
        n = len(nums)

        # count frequency of each number
        for num in nums:
            if num in empty_dict:
                empty_dict[num] += 1
            else:
                empty_dict[num] = 1

        # sort the keys
        sorted_keys = sorted(empty_dict.keys())

        # map each number to how many are smaller than it
        smaller_count = {}
        count = 0
        for key in sorted_keys:
            smaller_count[key] = count
            count += empty_dict[key]

        # build the answer list
        for num in nums:
            empty_list.append(smaller_count[num])

        return empty_list
```

```python
#https://leetcode.com/problems/check-if-the-sentence-is-pangram/
class Solution:
    def checkIfPangram(self, sentence: str) -> bool:
        if(len(sentence) < 26):
            return False

        for i in range(26):
            if chr(97+i) not in sentence:
                return False
        return True
```

```python
#https://leetcode.com/problems/count-items-matching-a-rule/
class Solution:
    def countMatches(self, items: List[List[str]], ruleKey: str, ruleValue: str) -> int:
        if ruleKey == "type":
            rulenum = 0
        elif ruleKey == "color":
            rulenum = 1
        else:
            rulenum = 2

        count = 0
        for i in range(len(items)):
            if items[i][rulenum] == ruleValue:
                count += 1

        return count
```

```python
#https://leetcode.com/problems/reverse-string/
class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        left, right = 0, len(s) - 1
        while left < right:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
```

Given a string that may contain extra spaces, find the length of the last word. A word is defined as a sequence of non-space characters. The solution ignores trailing spaces and counts characters of the final word by scanning from the end.

```python
#https://leetcode.com/problems/length-of-last-word/
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        end = len(s) - 1
        while s[end] == " ":
            end -= 1

        start = end
        while start >= 0 and s[start] != " ":
            start -= 1

        return end - start
```

```python
#https://leetcode.com/problems/sorting-the-sentence/
class Solution:
    def sortSentence(self, s: str) -> str:
        words = s.split()
        # create empty list of correct size
        result = [None] * len(words)
        for word in words:
            # Convert the last character (digit) to a 0-based index
            # ord('1') = 49 → 49 - 49 = 0
            # ord('2') = 50 → 50 - 49 = 1
            index = ord(word[-1]) - 49
            clean_word = word[:-1]    # remove last char
            result[index] = clean_word

        return " ".join(result)
```

```python
#https://leetcode.com/problems/split-a-string-in-balanced-strings/
class Solution:
    def balancedStringSplit(self, s: str) -> int:
        cnt_R = 0
        cnt_L = 0
        cnt = 0
        for c in s:
            if c == "R":
                cnt_R += 1
            elif c == "L":
                cnt_L += 1

            if cnt_R == cnt_L:
                cnt += 1
                cnt_R = 0
                cnt_L = 0

        return cnt
```

```python
#https://leetcode.com/problems/first-unique-character-in-a-string/
class Solution:
    def firstUniqChar(self, s: str) -> int:
        count = {}
        for char in s:
            if char in count:
                count[char] += 1
            else:
                count[char] = 1

        for i in range(len(s)):
            if count[s[i]] == 1:
                return i

        return -1
```

```python
#https://leetcode.com/problems/find-the-difference/
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        count = {}

        # Count characters in string t
        for c in t:
            if c in count:
                count[c] += 1
            else:
                count[c] = 1

        # Subtract counts for characters in string s
        for c in s:
            count[c] -= 1
            if count[c] == 0:
                del count[c]

        # Converts the dictionary keys into a list.
        # Returns the first key in the dictionary
        return list(count.keys())[0]
```

```python
#https://leetcode.com/problems/number-of-good-pairs/
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        count = {}

        for n in nums:
            if n in count:
                count[n] += 1
            else:
                count[n] = 1

        good_pair = 0

        for value in count.values():  # iterate over values only
            if value > 1:
                good_pair += value * (value - 1) // 2


        return good_pair
```

```python
#https://leetcode.com/problems/count-the-number-of-consistent-strings/
class Solution:
    def countConsistentStrings(self, allowed: str, words: List[str]) -> int:
        count = 0
        for w in words:                 # iterate over each word
            valid = True                # assume word is valid
            for c in w:                 # check each character
                if c not in allowed:
                    valid = False       # word is invalid
                    break               # stop checking this word

            if valid:
                count += 1              # count valid word

        return count
```

⚠️ Dict iteration note

*   dict.items() → gives (key, value) tuples
*   dict.keys() → gives keys only


*   dict.values() → gives values only

Mistake: comparing a tuple from .items() with an int → TypeError

Rule: if you need only the values, always use .values()

**Full solution**

```python
from collections import deque
from typing import List

class Solution:
    def latestDayToCross(self, row: int, col: int, cells: List[List[int]]) -> int:

        # Helper: convert 1-based to 0-based
        def convert_to_grid(r, c):
            return r - 1, c - 1

        # Helper: check if a path exists using BFS
        def is_path_exists(grid):
            directions = [(-1,0), (1,0), (0,-1), (0,1)]
            queue = deque()
            visited = [[False]*col for _ in range(row)]

            # Start BFS from all land cells in top row
            for c_idx in range(col):
                if grid[0][c_idx] == 0:
                    queue.append((0, c_idx))
                    visited[0][c_idx] = True

            while queue:
                r_idx, c_idx = queue.popleft()
                if r_idx == row - 1:  # reached bottom
                    return True
                for dr, dc in directions:
                    nr, nc = r_idx + dr, c_idx + dc
                    if 0 <= nr < row and 0 <= nc < col and grid[nr][nc] == 0 and not visited[nr][nc]:
                        queue.append((nr, nc))
                        visited[nr][nc] = True
            return False

        # Helper: flood the grid up to 'day' days
        def flood_grid(day):
            grid = [[0]*col for _ in range(row)]
            for i in range(day):
                r, c = convert_to_grid(cells[i][0], cells[i][1])
                grid[r][c] = 1
            return grid

        # Binary search for the last day
        left, right = 0, len(cells)
        last_day = 0

        while left <= right:
            mid = (left + right) // 2
            grid = flood_grid(mid)

            if is_path_exists(grid):
                last_day = mid
                left = mid + 1
            else:
                right = mid - 1

        return last_day
```

**Full Solution**

```python
class Solution:
    def sumFourDivisors(self, nums: List[int]) -> int:
        # store the cumulative sum of all divisors of numbers in nums that have exactly 4 divisors.
        total = 0
        for n in nums:
            cnt = 0 # counts how many divisors n has
            sum_div = 0 # stores the sum of these divisors
            root = int(n**0.5) # the integer square root of n
            # We only need to check up to the square root because divisors come in pairs (i, n // i)

            # Loop from 1 up to root
            # i is a candidate divisor of n
            for i in range (1, root + 1):
                if n % i == 0:
                    j = n // i

                    # Special case: if i equals j (this happens when n is a perfect square), we only count it once
                    # Example: for n = 4, i = 2, j = 2, we add 2 to the sum and increment count by 1.
                    if i == j:
                        cnt += 1
                        sum_div += 1
                    # If i != j, then we have two distinct divisors: i and j
                    # Add both to the sum and increment the count by 2.
                    else:
                        cnt += 2
                        sum_div += i + j
                    if cnt > 4:
                        break

            if cnt == 4:
                total += sum_div

        return total
```

```python
#https://leetcode.com/problems/two-sum/
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        dictionary = {}
        for i in range(n):
            if(target - nums[i]) in dictionary:
                return [dictionary[target - nums[i]], i]
            dictionary[nums[i]] = i
```

```python
#https://leetcode.com/problems/concatenation-of-array/
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        n = len(nums)
        empty_list = []
        for i in range(n):
            empty_list.append(nums[i])
        for i in range(n):
            empty_list.append(nums[i])

        return empty_list
```

```python
#https://leetcode.com/problems/running-sum-of-1d-array/
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        empty_list = []
        n = len(nums)
        summa = 0
        for i in range(n):
            summa = summa + nums[i]
            empty_list.append(summa)

        return empty_list
```

```python
#https://leetcode.com/problems/richest-customer-wealth/
class Solution:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        empty_list = []
        rows = len(accounts)
        cols = len(accounts[0])
        for i in range(rows):
            summa = 0
            for j in range(cols):
                summa = summa + accounts[i][j]
            empty_list.append(summa)
        maxi = 0
        for i in range(len(empty_list)):
            if empty_list[i] > maxi:
                maxi = empty_list[i]
        return maxi
```

```python
#https://leetcode.com/problems/kids-with-the-greatest-number-of-candies/
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        n = len(candies)
        maxi = 0
        empty_list = []
        for i in range(n):
            if candies[i] > maxi:
                maxi = candies[i]

        for i in range(n):
            if candies[i] + extraCandies >= maxi:
                empty_list.append(True)
            else:
                empty_list.append(False)

        return empty_list
```

```python
#https://leetcode.com/problems/shuffle-the-array/
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        empty_list = []
        n = len(nums) // 2
        for i in range(n):
            empty_list.append(nums[i])
            empty_list.append(nums[i+n])

        return empty_list
```

```python
#https://leetcode.com/problems/build-array-from-permutation/
class Solution:
    def buildArray(self, nums: List[int]) -> List[int]:
        return [nums[num] for num in nums]
```

```python
#https://leetcode.com/problems/how-many-numbers-are-smaller-than-the-current-number/
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        empty_dict = {}
        empty_list = []
        n = len(nums)

        # count frequency of each number
        for num in nums:
            if num in empty_dict:
                empty_dict[num] += 1
            else:
                empty_dict[num] = 1

        # sort the keys
        sorted_keys = sorted(empty_dict.keys())

        # map each number to how many are smaller than it
        smaller_count = {}
        count = 0
        for key in sorted_keys:
            smaller_count[key] = count
            count += empty_dict[key]

        # build the answer list
        for num in nums:
            empty_list.append(smaller_count[num])

        return empty_list
```

```python
#https://leetcode.com/problems/check-if-the-sentence-is-pangram/
class Solution:
    def checkIfPangram(self, sentence: str) -> bool:
        if(len(sentence) < 26):
            return False

        for i in range(26):
            if chr(97+i) not in sentence:
                return False
        return True
```

```python
#https://leetcode.com/problems/count-items-matching-a-rule/
class Solution:
    def countMatches(self, items: List[List[str]], ruleKey: str, ruleValue: str) -> int:
        if ruleKey == "type":
            rulenum = 0
        elif ruleKey == "color":
            rulenum = 1
        else:
            rulenum = 2

        count = 0
        for i in range(len(items)):
            if items[i][rulenum] == ruleValue:
                count += 1

        return count
```

```python
#https://leetcode.com/problems/reverse-string/
class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        left, right = 0, len(s) - 1
        while left < right:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
```

```python
#https://leetcode.com/problems/defanging-an-ip-address/
class Solution:
    def defangIPaddr(self, address: str) -> str:
        s = ""
        for i in address:
            if i == ".":
                s += "[.]"
            else:
                s += i
        return s
```

```python
#https://leetcode.com/problems/valid-palindrome/
class Solution:
    def isPalindrome(self, s: str) -> bool:
        clean = ""
        for ch in s:
            if ch.isalnum():
                clean += ch.lower()

        return clean == clean[::-1]
```

```python
#https://leetcode.com/problems/jewels-and-stones/
class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        cnt = 0
        for ch in stones:
            if ch  in jewels:
                cnt += 1
        return cnt
```

```python
#https://leetcode.com/problems/maximum-number-of-words-found-in-sentences/
class Solution:
    def mostWordsFound(self, ss: List[str]) -> int:
        return max(s.count(" ") for s in ss) + 1
```

*the Two-Pointer technique*

Given a string that may contain extra spaces, find the length of the last word. A word is defined as a sequence of non-space characters. The solution ignores trailing spaces and counts characters of the final word by scanning from the end.

```python
#https://leetcode.com/problems/length-of-last-word/
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        end = len(s) - 1
        while s[end] == " ":
            end -= 1

        start = end
        while start >= 0 and s[start] != " ":
            start -= 1

        return end - start
```

**Full solution**

```python
from collections import deque
from typing import List

class Solution:
    def latestDayToCross(self, row: int, col: int, cells: List[List[int]]) -> int:

        # Helper: convert 1-based to 0-based
        def convert_to_grid(r, c):
            return r - 1, c - 1

        # Helper: check if a path exists using BFS
        def is_path_exists(grid):
            directions = [(-1,0), (1,0), (0,-1), (0,1)]
            queue = deque()
            visited = [[False]*col for _ in range(row)]

            # Start BFS from all land cells in top row
            for c_idx in range(col):
                if grid[0][c_idx] == 0:
                    queue.append((0, c_idx))
                    visited[0][c_idx] = True

            while queue:
                r_idx, c_idx = queue.popleft()
                if r_idx == row - 1:  # reached bottom
                    return True
                for dr, dc in directions:
                    nr, nc = r_idx + dr, c_idx + dc
                    if 0 <= nr < row and 0 <= nc < col and grid[nr][nc] == 0 and not visited[nr][nc]:
                        queue.append((nr, nc))
                        visited[nr][nc] = True
            return False

        # Helper: flood the grid up to 'day' days
        def flood_grid(day):
            grid = [[0]*col for _ in range(row)]
            for i in range(day):
                r, c = convert_to_grid(cells[i][0], cells[i][1])
                grid[r][c] = 1
            return grid

        # Binary search for the last day
        left, right = 0, len(cells)
        last_day = 0

        while left <= right:
            mid = (left + right) // 2
            grid = flood_grid(mid)

            if is_path_exists(grid):
                last_day = mid
                left = mid + 1
            else:
                right = mid - 1

        return last_day
```
